```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
unique_items = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
number_of_purchase = len(purchase_data)
total_revenue = (number_of_purchase * average_price)
average_price = "${:.2f}".format(average_price)
total_revenue = "${:.2f}".format(total_revenue)
purchaing_analysis = pd.DataFrame({"Number of Unique Items" : [unique_items] , "Average Price" : [average_price],
                                            "Number of Purchases" : [number_of_purchase], "Total Revenue" : [total_revenue]})
purchaing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
pd.options.display.float_format = '{:,.2f}%'.format

wedves = purchase_data.drop_duplicates('SN')
#wedves = wedves.drop_duplicates('SN')

gender_demographics = wedves['SN'].groupby(wedves['Gender']).count().to_frame()
gender_demographics = gender_demographics.rename(columns = {'SN' : 'Total Count'})
gender_demographics['Percentage of Players'] = (gender_demographics['Total Count']/len(wedves))*100
gender_demographics = gender_demographics.sort_values(by = ['Total Count'], ascending=False)
gender_demographics.index.name = None

gender_demographics
```

```python
pd.options.display.float_format = '${:,.2f}'.format

purchase_analysis = purchase_data['SN'].groupby(purchase_data['Gender']).count().to_frame()
purchase_analysis = purchase_analysis.rename(columns = {'SN' : 'Purchase Count'})
purchase_analysis['Average Purchase Price'] = purchase_data['Price'].groupby(purchase_data['Gender']).mean()
purchase_analysis['Total Purchase Value'] = purchase_analysis['Purchase Count'] * purchase_analysis['Average Purchase Price']
purchase_analysis['Avg. Total Purchase per Person'] = purchase_analysis['Total Purchase Value']/wedves['SN'].groupby(
    wedves['Gender']).count()

purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
pd.options.display.float_format = '{:,.2f}%'.format

bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

age_demo = wedves["SN"].groupby(pd.cut(wedves["Age"], bins, labels = group_names)).count().to_frame()
age_demo = age_demo.rename(columns = {'SN' : 'Total Count'})
age_demo['Percentage of Players'] = (age_demo['Total Count']/len(wedves))*100

age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pd.options.display.float_format = '${:,.2f}'.format

age_analysis = purchase_data["SN"].groupby(pd.cut(purchase_data["Age"], bins, labels = group_names)).count().to_frame()
age_analysis = age_analysis.rename(columns = {'SN' : 'Purchase Count'})
age_analysis['Average Purchase Price'] = purchase_data["Price"].groupby(
    pd.cut(purchase_data["Age"], bins, labels = group_names)).sum()/age_analysis['Purchase Count']
age_analysis['Total Purchase Value'] = age_analysis['Purchase Count'] * age_analysis['Average Purchase Price']
age_analysis['Avg. Total Purchase per Person'] = purchase_data["Price"].groupby(
    pd.cut(purchase_data["Age"], bins, labels = group_names)).sum()/age_demo['Total Count']

age_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Top_Spenders = purchase_data['SN'].value_counts().to_frame()
Top_Spenders = Top_Spenders.rename(columns = {'SN' : 'Purchase Count'})
Top_Spenders["Average Purchase Price"] = purchase_data['Price'].groupby(purchase_data['SN']).sum()/Top_Spenders["Purchase Count"]
Top_Spenders["Total Purchase Value"] = Top_Spenders['Average Purchase Price'] * Top_Spenders["Purchase Count"]
Top_Spenders = Top_Spenders.rename_axis("SN")

Top_Spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pd.options.display.float_format = '${:,.2f}'.format

items_count = purchase_data['Age'].groupby(purchase_data['Item ID']).count().to_frame()

popular_items = purchase_data[["Item ID","Price","Item Name"]].drop_duplicates('Item ID')
popular_items = pd.merge(items_count,popular_items,on='Item ID',how='outer')
popular_items = popular_items.rename(columns = {'Age' : 'Purchase Count' , 'Price' : 'Item Price'})
popular_items['Total Purchase Value'] = popular_items['Purchase Count'] * popular_items['Item Price']
popular_items = popular_items.sort_values(by='Purchase Count',ascending=False)
popular_items.set_index(['Item ID','Item Name'], inplace=True)
popular_items.head()
```
